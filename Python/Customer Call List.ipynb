```python
# Limpiaremos y estandarizaremos la siguiente lista de clientes para un area de servicio, el cual debe llamar solo a los clientes que 
# especifiquen ser contactados. 
```

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\valer\Downloads\Customer Call List.xlsx")
df
```

```python
#Eliminamos columnas que no son necesarias
df = df.drop(columns = "Not_Useful_Column")
```

```python
#Eliminamos duplicados
df = df.drop_duplicates()
df
```

```python
#Limpiamos las columnas de nombre y apellido que contienen caracteres distinto a letras
df["Last_Name"] = df["Last_Name"].str.strip("._/123")
df["First_Name"] = df["First_Name"].str.strip("._/123")
df
```

```python
# Estandarizamos el numero de telefono utilizando funcion regex convirtiendo a string y luego aplicando una funcion lambda para reescribir el numero de telefono en formato estandar

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df['Phone_Number'] = df['Phone_Number'].astype(str)
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
# Estandarizamos los valores nan-- y Na-- a en blanco

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
```

```python
df
```

```python
# Ordenamos la columna de Address, para que quede en 3 columnas distinta, para facilitar la lectura y búsqueda (calle, estado y zip)

df[["Street", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2, expand=True)
```

```python
df
```

```python
#Estandarizamos los valores de la columna, en una sola letra Y por si y N por No

df["Paying Customer"] = df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No","N")
```

```python
df
```

```python
# Realizamos la misma estandarizacion para la columna de contacto, la cual más adelante debemos filtrar
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes","Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No","N")
```

```python
df
```

```python
#Eliminamos la columna Address que ya no es necesaria
df = df.drop(columns = "Address")
```

```python
df
```

```python
# Estandarizamos los valores generales de la tabla de na a vacio
df = df.fillna('')
```

```python
# Estandarizamos los valores generales de la tabla de N/a a vacio
df = df.replace('N/a','')
```

```python
df
```

```python
# Filtramos las filas, que tienen la opción de No contactar

for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)

df
```

```python
# y también filtramos las filas, que no tienen numero de telefono que no sirven
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x, inplace = True)

df
```

```python
# Finalmente reseteamos el indice, para una lectura ordenada
df.reset_index(drop=True)
```
